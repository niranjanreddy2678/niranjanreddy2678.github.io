```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

```python
stopwords = list(STOP_WORDS)
nlp = spacy.load('en_core_web_sm')
```

```python
text = """Due to increase of the population, the usage of the internet resources wasn't the only thing increased but also hacking. Now, as a result the security is the major concern that needed to be taken care of. Intrusion detection helps in the maintaining and saving the system from both the internal and external influences by using of the genetic algorithm, artificial immune system and network-based protection algorithms. At some cases the rough set theory also plays a major role in providing the security. In our system we make analogy between intrusion detection system and artificial immune system, these help us to achieve our goal. The power system reconfiguration is a challenging task that it has to reconfigure its settings on its own without the manual intervention. At the distributive level the self-restorative actions and automatic corrective algorithms will come under the self-healing schemes. In this paper the author explains about the automatic repairing system using an AIS tool as an optimization technique to resolve the problems occurred in the power source and also taking into consideration the faults that are going to occur during the inner most switches which handles brake system. The self-repairing method using AIS is applied while reconfiguring the systems which are fault. In the collaborative filtering, it gives idea of having the internal communication between the two drones or two sets. Even the collaborative filtering may act as an antigen to an immune network which helps in communicating it with the other drones by the secure channel. This is the ongoing research and many experiments are yet to be performed. Most of the systems had adopted biological behavior in the architecture-based computing which is used in the traditional days in which certain numbers of tasks execute in parallel where there is no centralized system control. Lot of researchers have started computing the fitness levels of various computing techniques for determining unique features of AIS while trying to adopt to old computing architecture and few of them consider that applying these algorithms practically would really solve the real-world problems. In this paper the author conveyed the efficiency of handling datasets which contains real time data can be improved more by bringing out the most invaluable or hidden features of the architectures. """
```

```python
fil = nlp(text)
tokens = [token.text for token in fil]
```

```python
##finding the word freqencies
word_frequencies ={}
for word in fil:
    if word.text.lower() not in stopwords:
        if word.text.lower() not in punctuation:
            if word.text not in word_frequencies.keys():
                word_frequencies[word.text] = 1
            else:
                word_frequencies[word.text] += 1
```

```python
print(word_frequencies)
```

```python
max_frequencies = max(word_frequencies.values())
max_frequencies
```

```python
## Normalising the freqencies for the usage of summary
for word in word_frequencies.keys():
    word_frequencies[word] = word_frequencies[word]/max_frequencies

print(word_frequencies)
```

```python
sentence_tokens = [sent for sent in fil.sents]
```

```python
sentence_scores = {}
for sent in sentence_tokens:
        for word in sent:
            if word.text.lower() in word_frequencies.keys():
                if sent not in sentence_scores.keys():
                    sentence_scores[sent]=word_frequencies[word.text.lower()]
                else:
                    sentence_scores[sent]=word_frequencies[word.text.lower()]
```

```python
#here we are finding the scores for the sentences which are most repeated and  which are not and based on that we summarize
sentence_scores
```

```python
from heapq import nlargest
select_length = int(len(sentence_tokens)*0.25)
select_length
```

```python
summary = nlargest(select_length, sentence_scores, key = sentence_scores.get)
summary
```

```python
final_summary = [word.text for word in summary]
```

```python
summary = ' '.join(final_summary)
```

```python
print(summary)
```

## Major differences between nltk and spaCy i noticed are
###### NLTK supports various languages whereas spaCy have statistical models for 7 languages (English, German, Spanish, French, Portuguese, Italian, and Dutch). It also supports named entities for multi language. NLTK is a string processing library. It takes strings as input and returns strings or lists of strings as output. Whereas, spaCy uses object-oriented approach. When we parse a text, spaCy returns document object whose words and sentences are objects themselves.
